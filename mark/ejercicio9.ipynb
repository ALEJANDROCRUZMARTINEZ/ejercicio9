# c. Realiza un diagrama de barras.

```python
marcas_clase = [1, 2, 3, 4, 5, 6, 7, 8]
frecuencias = [7912, 3736, 1370, 1091, 1016, 445, 411, 405]
marcas_texto = ["Tsuru", "Pick up", "Sentra", "Jetta 4a, Generación", "Bora", "Estaquitas 35 Ton", "Pointer", "Sedan"]

plt.figure(figsize = (12, 6))

plt.barh(marcas_clase, frecuencias,
        height = 0.75, edgecolor = "k",
       color = ["#F9F3AC", "#DEF5A9", "#86F7B4", "#86ECF7", "#86AFF7", "#7A7BF5", "#C87AF5", "#EB7AF5"])

plt.yticks(marcas_clase, marcas_texto, fontsize = 15)
plt.xlabel("Número de unidades", fontsize = 20)
plt.ylabel("Tipos", fontsize = 20)
plt.title("Vehículos con mayor número de robos en enero - julio 2010 en el país", fontsize = 25)
plt.grid()
plt.show()
```

# Propuestas para reducir el robo de autos

- Estacionamiento seguro: Cuando estaciones tu vehículo, elige cuidadosamente el lugar. Opta por áreas bien iluminadas y con presencia de cámaras de seguridad.
- Medidas de seguridad adicionales: Además de las medidas de seguridad incorporadas en tu vehículo, considera la instalación de dispositivos adicionales de protección contra robos
- Buenas prácticas de seguridad: Adopta hábitos de seguridad sólidos al dejar tu vehículo desatendido. Siempre cierra y asegura todas las puertas y ventanas, incluso si solo te alejas por un corto período de tiempo.
